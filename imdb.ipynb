```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time
```

```python


driver = webdriver.Chrome()
wait = WebDriverWait(driver, 10)

driver.get(
    "https://www.imdb.com/search/title/?title_type=feature&release_date=2024-01-01,2024-12-31&genres=adventure"
)

time.sleep(5)

# ---------- CLICK "50 MORE" UNTIL END ----------
while True:
    try:
        load_more = wait.until(
            EC.element_to_be_clickable(
                (By.CSS_SELECTOR, "button.ipc-see-more__button")
            )
        )

        driver.execute_script(
            "arguments[0].scrollIntoView({block: 'center'});",
            load_more
        )
        time.sleep(1)
        load_more.click()
        time.sleep(3)

    except:
        print("Reached last page ")
        break

# ---------- SCRAPE MOVIES ----------
movies = []
movie_cards = driver.find_elements(
    By.CSS_SELECTOR, "li.ipc-metadata-list-summary-item"
)

print("Total movies loaded:", len(movie_cards))

for card in movie_cards:
    movie = {}

    # TITLE
    try:
        title_text = card.find_element(
            By.CSS_SELECTOR, "h3.ipc-title__text"
        ).text
        movie["title"] = title_text.split(". ", 1)[1]
    except:
        movie["title"] = "N/A"

    # DURATION
    try:
        metadata = card.find_elements(
            By.CSS_SELECTOR, "span.dli-title-metadata-item"
        )
        movie["duration"] = metadata[1].text if len(metadata) > 1 else "N/A"
    except:
        movie["duration"] = "N/A"

    # RATING
    try:
        movie["rating"] = card.find_element(
            By.CSS_SELECTOR, "span.ipc-rating-star--rating"
        ).text
    except:
        movie["rating"] = "N/A"

    #  VOTE COUNT (NEW)
    try:
        votes = card.find_element(
            By.CSS_SELECTOR, "span.ipc-rating-star--voteCount"
        ).text
        movie["vote_count"] = (
            votes.replace("(", "")
                 .replace(")", "")
                 .replace(",", "")
                 .strip()
        )
    except:
        movie["vote_count"] = "N/A"

    # STORYLINE
    try:
        movie["storyline"] = card.find_element(
            By.CSS_SELECTOR, "div.ipc-html-content-inner-div"
        ).text
    except:
        movie["storyline"] = "N/A"

    movies.append(movie)

driver.quit()

df = pd.DataFrame(movies)
print("Total movies scraped:", len(df))
print(df.head())
```

```python
df
```

```python
df.to_csv("imdb_adventure_movies_2024.csv", index=False)
print("CSV saved successfully!")
```
